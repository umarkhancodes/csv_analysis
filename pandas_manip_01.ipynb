```python
import pandas as pd
```

```python
A = 'Available in Eurpool'
B = 'Not Available in Europool'
C = 'Not Available in Goodyear'
```

```python
ep=pd.read_excel('EMEA_OE _Bid_Data2.xlsx')
ep_out=ep.copy()
ep_out["Status"]="NA"
ep_out=ep_out.astype(str)
ep_out.head()
```

```python
for index,row in ep_out.iterrows():
    cv=str(row['TireSize'])
    bp,ap=cv.split('.')
    if(ap==len(ap)*'0'):
        ep_out.at[index,'TireSize']=bp
    else:
        ep_out.at[index,'TireSize']=cv
ep_out.head()
```

```python
ep_ts=ep_out['TireSize'].unique()
```

```python
rel=pd.read_excel("EMEA_OE _Bid_Data1.xlsx")
rel.head()
```

```python

for index,row in rel.iterrows():
    ts=row['TireSize']
    if ts in ep_ts:
        ep_out.loc[(ep_out['TireSize']==ts),"Status"]=A
    else:
        ap_out=pd.DataFrame([['NA','NA','NA','NA','NA','NA',ts,B]],columns=list(ep_out.columns))
        ep_out=ep_out.append(ap_out,ignore_index=True)
```

```python
ep_out.loc[(ep_out['Status']=='NA'),'Status']=C
```
